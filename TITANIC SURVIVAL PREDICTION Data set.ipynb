#### Dropping Cabin column or Feature as out of 891 rows 687 are Missing Values which is a Very High number. So we can drop this column.

#### Age column has 177 Missing values. As age is Numeric type column(float), so, ther Missing values in this column can be replaced by median or mean. Embarked Column has 2 missing values and as it is a categorical column so the missing values in this column can be replaced by mode. 

```python
dft = dft.drop(["Cabin"], axis = 1)
dft
```

```python
median = dft['Age'].median()
dft['Age'].replace(np.nan, median, inplace=True)
dft.tail()
```

```python
mode = dft["Embarked"].mode().values[0]

dft["Embarked"] = dft["Embarked"].replace(np.nan,mode)
dft.tail()
```

#### checking one more time for the missing values.

```python
dft.info()
```

#### Also the columns Name and Ticket can be Dropped. As clearly it does not Logically impact or help in classifying the Survived and Non Surviived one.

```python
dft = dft.drop(["Name", "Ticket"], axis = 1)
dft
```

### One Hot Encoding

```python
dummies = pd.get_dummies(dft[["Sex", "Embarked", "Pclass"]], columns = ["Sex", "Embarked", "Pclass"], drop_first =True)
dummies
```

```python
columns = ["Sex", "Embarked", "Pclass"]
dft = pd.concat([dft, dummies], axis=1)
dft.drop(columns, axis=1, inplace=True)
```

```python
dft.head()
```

```python
dft.shape
```

#### Split the Dat

```python
x = dft.drop(['Survived'],axis=1)
y = dft[['Survived']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
```

# Logistic Regression

```python
reg_model = LogisticRegression()
reg_model.fit(x_train, y_train)
```

```python
reg_model.score(x_train, y_train)
```

```python
reg_model.score(x_test, y_test)
```

```python
predictions=reg_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

### Code to Import Metrics

```python
from sklearn import metrics
```

### Code for Classification Report 

```python
print(metrics.classification_report(y_test, predictions))
```

### Code to Import Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

### the challenge with this confusion matrix is that we do not know which value is TP, FP, FN, TN.

```python
cm=metrics.confusion_matrix(y_test, predictions, labels=[1, 0])

df_cm = pd.DataFrame(cm, index = [i for i in ["1", "0"]])
sns.heatmap(df_cm, annot = True, fmt = 'g')
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dTree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dTree.fit(x_train, y_train)
```

## Scoring our Decision Tree

```python
print(dTree.score(x_train, y_train))
print(dTree.score(x_test, y_test))
```

### Thwe above score of train and test data clearly tells us the its a overfitting model.

### Let's see how to control the overfitting

### Reducing Overfitting(Regularization)

```python
dTreeR = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3, random_state = 0)
dTreeR.fit(x_train, y_train)
print(dTreeR.score(x_train, y_train))
print(dTreeR.score(x_test, y_test))
```

# Ensemble Learning - Bagging

```python
from sklearn.ensemble import BaggingClassifier
bgcl = BaggingClassifier(n_estimators = 100, base_estimator = dTree, random_state = 0)
bgcl = bgcl.fit(x_train, y_train)
y_predict = bgcl.predict(x_test)
print(bgcl.score(x_train, y_train))
print(bgcl.score(x_test, y_test))
```

```python
from sklearn.ensemble import BaggingClassifier
bgcl = BaggingClassifier(n_estimators = 50, base_estimator = dTreeR, random_state = 0)
bgcl = bgcl.fit(x_train, y_train)
y_predict = bgcl.predict(x_test)
print(bgcl.score(x_train, y_train))
print(bgcl.score(x_test, y_test))
```

# Ensemble Learning - AdaBoosting

```python
from sklearn.ensemble import AdaBoostClassifier
abcl = AdaBoostClassifier(n_estimators=30, random_state=0)
abcl = abcl.fit(x_train, y_train)
y_predict = abcl.predict(x_test)
print(abcl.score(x_train, y_train))
print(abcl.score(x_test, y_test))
```

### As we can see here that score of train data is 0.86 and that of  test data is 0.79, which exceeds the max value of difference between train and test score which is equal to 5%. So the model is Overfitting. 

```python
cm=metrics.confusion_matrix(y_test, y_predict, labels=[0, 1])

df_cm = pd.DataFrame(cm, index = [i for i in ["No", "Yes"]],
                    columns = [i for i in ["No", "Yes"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True, fmt='g')
```
