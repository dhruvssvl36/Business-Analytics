### Dealing with Missing Values

```python
lr.describe()
```

#### .describe() command gives the information about the Stats of the Numerical columns and not the Categorical ones. So by writting this Horsepower column is not shown. 

```python
lr.describe(include="all")
```

####  .describe(include = "all") command gives the information about the Stats of Numerical as well as categorical column. Here the Horsepower column appeared. From here also we can know that there is something wrong with the Horsepower column. Also by running .info(), we can see that there are no missing values in the Horse power column. Then what is the problem with it? Why it is showing dtype as object(Categorical)?

#### On searching its csv file we can see that it has some "?" signs .

#### when we run the code lr.describe(), the horsepower column was not shown but when we ran the code lr,describe(include="all"), the horsepower column was shown. Implies it has been considered as Categorical column. So something is wrong over there. So now when we run dtypes we can see that it is considered under object data type.

```python
lr.dtypes
```

#### That's because Horsepower column has Missing values and in place of missing values "?" is written.

```python
hpIsDigit=pd.DataFrame(lr.horsepower.str.isdigit())# Here we are checking whether this string is made up of digits or not.horsepower column is being shown as object. If string is made up of digits store True. If not made up of digits then it will be storing false.
lr[hpIsDigit['horsepower'] == False]   # Display wherever there are false.
```

#### I checked whether the string is made up of digits or not. If it is made up of digits, it would be true, if it is not made of digits, then false.  So I wanted to see only the cases wherever the horsepower column is equal to false so, I wrote code so that we can only see the rows having "?". So here as we can see 6 rows are having "?" sign.

#### So now I am going to replace "?" by np.nan.

```python
lr=lr.replace('?',np.nan)
lr[hpIsDigit['horsepower'] == False]
```

```python
lr.dtypes
```

```python
lr.describe()
```

```python
lr.describe(include="all")
```

```python
lr.median()   # Median of the columns.
```

#### Here I am Replacing null or missing values with the median values. Also I am converting the data type of horsepower column from object to float because now as we replaced the "?" by np.nan and np.nan by median, so now this column contains only the numerical values(float). 

```python
# replace the missing values with median value.
# Note, we do not need to specify the column names below
# Every column's missing value is replaced with that column's median respectively.


medianFiller = lambda x: x.fillna(x.median())
lr = lr.apply(medianFiller,axis=0)

lr['horsepower'] = lr['horsepower'].astype('float64')
```

# Split Data

```python
# Let's build our Linear Model.
# Independent Variables.
x = lr.drop(['Kilometer_per_liter'],axis=1)
# The Dependent variable
y = lr[['Kilometer_per_liter']]
```

### Split x and y into Training and Test data set in 70:30 Ratio

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

# Linear Regression

### Fit Linear Model

```python
reg_model = LinearRegression()   #Instanciating model
reg_model.fit(x_train, y_train)
```

```python
reg_model.score(x_train, y_train)
```

```python
reg_model.score(x_test, y_test)
```

# Polynomial Regression

#### Curve Fitting by Importing Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

poly = PolynomialFeatures(degree=2, interaction_only=True)
x_train2 = poly.fit_transform(x_train)
x_test2 = poly.fit_transform(x_test)
# Here we are doing Transformation only on the x data i.e. Independent Attributes. So only on the independent attributes we are doing transformation.
poly_clf = linear_model.LinearRegression()

poly_clf.fit(x_train2, y_train)

y_pred = poly_clf.predict(x_test2)

#print(y_pred)

# In samplr R^2 will always improve with the number of variables!
print(poly_clf.score(x_train2, y_train))
```

```python
#Out off sample(testing) R^2 is our measure of sucess and does improve
print(poly_clf.score(x_test2, y_test))
```
